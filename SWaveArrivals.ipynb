```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Specify the parent folder 
BASE_DIRECTORY = '/Users/anna/Desktop/Svels (2)/'

# Specify the folder 
fname0 = "1A_S_1V_amp_1MHz_100ave"
dat = np.load(BASE_DIRECTORY + fname0 + ".npy")

# Plot
plt.plot(dat[:, 0], dat[:, 1])
plt.grid(True)

plt.axvline(x=0.0000175, color='red', linestyle='-', label='Picked')

plt.xlabel("Time (us)")
plt.ylabel("Voltage (Vpp)")
plt.title("S-wave Sample 1A_SS Voltage vs Time")

plt.legend()

# Save 
#plt.savefig('/Users/anna/Desktop/1AVT.png', dpi=300, bbox_inches='tight')


# Show the plot
plt.show()
```
