```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Specify the parent folder 
parent_folder = '/Users/anna/Desktop/MastersData/SeismicWaveVelocity'

# Specify the folder names containing the .npy files
folder_names = ['PwaveTrial1', 'PwaveTrial2', 'PwaveTrial3', 'SwaveTrial1', 'SwaveTrial2', 'SwaveTrial3']

# Iterate over each folder
for folder_name in folder_names:
    # Construct the full path of the current folder
    folder_path = os.path.join(parent_folder, folder_name)
    
    # Get a list of all files in the current folder
    file_list = os.listdir(folder_path)
    
    # Iterate over each file in the folder
    for file_name in file_list:
        # Check if the file is a .npy file
        if file_name.endswith('.npy'):
            # Load the .npy file using np.load
            file_path = os.path.join(folder_path, file_name)
            array_data = np.load(file_path)
            
            # Reshape the array to (2, -1)
            reshaped_array = np.reshape(array_data, (2, -1))
            
            # Save the reshaped array back to the file
            np.save(file_path, reshaped_array)
            


```

```python
# Define the function to load and reshape the .npy file
def load_and_reshape(file_path):
    # Load the .npy file using np.load
    array_data = np.load(file_path)
    
    # Reshape the array to (2, -1)
    return np.reshape(array_data, (2, -1))

# Specify the parent folder 
parent_folder = '/Users/anna/Desktop/MastersData/SeismicWaveVelocity/PwaveTrial1'

# File names to plot
file_names = ['1A_LS_1FHMZ_A10-Ptransducer.npy', '1A_SS_1FHMZ_A10-Ptransducer.npy']

# Iterate over each file name
for i, file_name in enumerate(file_names):
    # Construct the full path of the file
    file_path = os.path.join(parent_folder, file_name)
    
    # Load and reshape the data
    reshaped_data = load_and_reshape(file_path)
    
    # Determine the title based on the file name
    title = '1A_LS' if 'LS' in file_name else '1A_SS'
    
    # Plot the data
    plt.figure()
    plt.plot(reshaped_data[0], reshaped_data[1])
    plt.title(f'P wave Sample {title} Voltage vs Time Trial 1', fontsize=14, fontweight='bold')
    plt.xlabel('Time (us)', fontsize=15, fontweight='bold')
    plt.ylabel('Voltage (Vpp)', fontsize=15, fontweight='bold')

    # Make x and y tick labels bold
    plt.xticks(fontsize=12, fontweight='bold')
    plt.yticks(fontsize=12, fontweight='bold')
    
    # Set the common vertical line for both titles
    if title == '1A_LS':
        plt.axvline(x=0.000025, color='red', linestyle='-', linewidth=0.5)
    
    # Add vertical lines specific to each title
    if title == '1A_SS':
        plt.axvline(x=0.0000093, color='red', linestyle='-', linewidth=0.5)
    
    plt.xlim(0,0.000055)
    plt.grid(True)
    plt.show()

```
