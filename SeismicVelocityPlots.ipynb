```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import matplotlib.lines as mlines
```

```python
#P-wave LS Velocities 
# Data
samples = ['5A', '5B','2A', '2B','1A', '1B', '6B', '6A', '4A']
velocities = [5.587510165, 5.569234403, 5.662, 5.429573199,
              5.094422486, 5.157513228, 4.491429952, 3.707083986, 3.5597563]
percentages = [0.24, 0.35, 0.43, 0.55, 0.86, 0.90, 1.80, 1.83, 3.18]
std_devs = [0.396025, 0.373484, 0.000000, 0.068971, 0.812899,
            0.083487, 0.389522, 0.110654, 0.244236]

# Define highlight colors
highlight_samples = ['1A', '1B', '2A', '2B']
colors = ['red' if sample in highlight_samples else 'blue' for sample in samples]

# Convert to numpy arrays
x = np.array(percentages)
y = np.array(velocities)

# Calculate trendline
coeffs = np.polyfit(x, y, deg=1)
trendline = np.poly1d(coeffs)
x_vals = np.linspace(min(x), max(x), 100)
y_vals = trendline(x_vals)

# Calculate correlation and R²
r = np.corrcoef(x, y)[0, 1]
r_squared = r**2

# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, y, yerr=std_devs, fmt='o', color='black', ecolor='gray', capsize=5, markersize=0.5)

# Plot data points
for i, sample in enumerate(samples):
    plt.scatter(x[i], y[i], color=colors[i], s=200, edgecolor='k')

# Plot trendline
trendline_plot, = plt.plot(x_vals, y_vals, color='gray', linestyle='--', linewidth=2)

# Labels and title
plt.ylabel('Average Wave Velocity (km/s)', fontsize=14)
plt.xlabel('Percentages of Carbonate Minerology', fontsize=14)
plt.title('P-Wave Velocities (LS)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Legend handles
highlight_legend = mlines.Line2D([], [], color='red', marker='o', linestyle='',
                                 markersize=10, label='BOIC')
default_legend = mlines.Line2D([], [], color='blue', marker='o', linestyle='',
                               markersize=10, label='Cedars')
trendline_legend = mlines.Line2D([], [], color='gray', linestyle='--', linewidth=2,
                                 label=f'Trendline\nr = {r:.2f}, R² = {r_squared:.2f}')

# Add legend
plt.legend(handles=[highlight_legend, default_legend, trendline_legend],
           title='Sample Sites', title_fontsize=13, fontsize=11, loc='upper right')

# Annotate sample names
for i, sample in enumerate(samples):
    plt.text(x[i], y[i] + 0.15, sample, fontsize=11, ha='center', weight='bold')

plt.tight_layout()
plt.savefig('PP-LongSide.png')
plt.show()

```
