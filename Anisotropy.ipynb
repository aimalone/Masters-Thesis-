```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

# Data
samples = ['5A', '5B', '2A', '2B', '1A', '1B', '6B', '6A', '4A']
densities = np.array([2.95, 3.01, 2.68, 2.62, 2.61, 2.61, 2.49, 2.34, 2.11])
terms_pB = np.array([17.80890483, 10.64218707, 6.592535362, 2.160063122,
                     5.721029415, 2.035363251, 1.253916495, 20.01817464, 9.437480219])
terms_sB = np.array([19.67702419, 14.27, 7.551611187, 2.544087847,
                     7.839297272, 1.972924522, 2.975833198, 16.92934187, 4.61000944])

# Samples to be colored blue and red
blue_samples = ['5A', '5B', '6B', '6A', '4A']
red_samples = ['2A', '2B', '1A', '1B']

# --- Sort everything by density ---
data = sorted(zip(densities, samples, terms_pB, terms_sB), key=lambda t: t[0])
dens_sorted, sam_sorted, p_sorted, s_sorted = zip(*data)
dens_sorted = np.array(dens_sorted)
p_sorted = np.array(p_sorted)
s_sorted = np.array(s_sorted)

# --- Compute R manually ---
def compute_r2(x, y):
    y_pred = np.polyval(np.polyfit(x, y, 1), x)
    ss_res = np.sum((y - y_pred)**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    return 1 - ss_res/ss_tot

r2_p = compute_r2(dens_sorted, p_sorted)
r2_s = compute_r2(dens_sorted, s_sorted)

print(f"R (P-wave) = {r2_p:.3f}")
print(f"R (S-wave) = {r2_s:.3f}")

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot for P-wave and S-wave
for x, sample, p_val, s_val in zip(dens_sorted, sam_sorted, p_sorted, s_sorted):
    color = 'red' if sample in red_samples else 'blue'
    plt.scatter(x, p_val, color=color, marker='x', s=100, edgecolor='black', linewidths=1.5)
    plt.scatter(x, s_val, color=color, marker='s', s=100, edgecolor='black')

    # Put sample name above the point
    y_max = max(p_val, s_val)
    plt.text(x, y_max + 0.3, sample, ha='center', va='bottom', fontsize=9, fontweight='bold')

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.6)

# Axis labels & title
plt.xlabel('Density (g/cm)')
plt.ylabel('Anisotropy Values (%)')
plt.title('P- and S-Wave Anisotropies')

# X-axis ticks: show density values
xtick_labels = [f'{d:.2f}' for d in dens_sorted]
plt.xticks(dens_sorted, xtick_labels, fontsize=10)

# Small horizontal margins
plt.margins(x=0.05)

# Legend handles
handles = [
    Line2D([0], [0], marker='x', color='w', markerfacecolor='blue', markeredgecolor='black',
           markersize=10, label='Cedars P-wave'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='blue', markeredgecolor='black',
           markersize=10, label='Cedars S-wave'),
    Line2D([0], [0], marker='x', color='w', markerfacecolor='red', markeredgecolor='black',
           markersize=10, label='BOIC P-wave'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='red', markeredgecolor='black',
           markersize=10, label='BOIC S-wave')
]
plt.legend(handles=handles)

# Add R values on plot
plt.text(0.015, 0.75, f"R (P-wave) = {r2_p:.3f}\nR (S-wave) = {r2_s:.3f}",
         transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top',
         bbox=dict(facecolor='white', alpha=0.7, edgecolor='black'))

plt.tight_layout()
plt.savefig('aniRIGHT_density!!.png', dpi=300)
plt.show()
```
