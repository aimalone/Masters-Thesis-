```python
import numpy as np
import matplotlib.pyplot as plt

# Sample names
samples = ["1A", "1B", "2A", "2B", "4A", "5A", "5B", "6A", "6B"]

# Descriptive terms mapping to custom ranges
term_mapping = {
    'Isotropic': 1,   # < 2
    'Fairly': 4,      # 2-6 (midpoint)
    'Moderately': 13, # 6-20 (midpoint)
    'Highly': 30,     # 20-40 (midpoint)
    'N/A': np.nan     # Use NaN for missing values
}

# Descriptive Terms S and P
terms_s = [22.01061665, 11.17263947, 4.748702426, 2.160063119, 1.682265007, 17.80890483, 15.44231809, 27.00541554, 37.00311328]
terms_p = [16.31822319, 6.736445908, 28.22807945, 5.153733431, 18.94831987, 19.67702423, 15.92062455, 18.99048875, 3.112866125]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the S-Wave and P-Wave values
plt.scatter(samples, terms_s, label='S-Wave', color='limegreen', marker='o', s=100)  # Bigger markers
plt.scatter(samples, terms_p, label='P-Wave', color='purple', marker='x', s=100)   # Bigger markers

# Set x-axis and y-axis labels
plt.xlabel('Sample Names', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Add grid lines for both Y and X axes
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add a legend
plt.legend(loc='center left', bbox_to_anchor=(1, 0.95))

# Add title
plt.title('Values of P and S Waves by Sample', fontsize=14)

# Annotate each point with its value
for i, (s_value, p_value) in enumerate(zip(terms_s, terms_p)):
    plt.annotate(f'{s_value:.2f}', (samples[i], s_value), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)
    plt.annotate(f'{p_value:.2f}', (samples[i], p_value), textcoords="offset points", xytext=(0,-10), ha='center', fontsize=9)

# Show the plot
plt.show()

# plt.savefig('anisotropy.png')

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Original data
percentages = [0.24, 0.35, 0.43, 0.55, 0.86, 0.90, 1.80, 1.83, 3.18]
y_values = [1.97, 0.30, 0.37, 0.54, 2.45, 4.69, 135.91, 195.97, 178.48]
datasets = ['5A', '5B', '2A', '2B', '1A', '1B', '6B', '6A', '4A']
colors = ['blue', 'blue', 'red', 'red', 'red', 'red', 'blue', 'blue', 'blue']

# Sort data for plotting
sorted_data = sorted(zip(percentages, y_values, datasets, colors))
sorted_percentages, sorted_y_values, sorted_datasets, sorted_colors = zip(*sorted_data)

# Calculate correlation coefficient (r) and R-squared (R²)
r_matrix = np.corrcoef(sorted_percentages, sorted_y_values)
r = r_matrix[0, 1]
R_squared = r**2
print(f"Correlation coefficient (r): {r:.4f}")
print(f"Coefficient of determination (R²): {R_squared:.4f}")

# Create scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(sorted_percentages, sorted_y_values, c=sorted_colors, edgecolor='black', s=100)

# Add labels above each point
for x, y, label in zip(sorted_percentages, sorted_y_values, sorted_datasets):
    plt.text(x, y + 5, label, ha='center', fontsize=10, weight='bold')

# Fit linear trendline using numpy
coeffs = np.polyfit(sorted_percentages, sorted_y_values, deg=1)
trendline = np.poly1d(coeffs)
x_vals = np.linspace(min(sorted_percentages), max(sorted_percentages), 100)
plt.plot(x_vals, trendline(x_vals), color='gray', linestyle='--', linewidth=2, label='Trendline')

# Labels and grid
plt.xlabel('Carbonate Mineralization (%)', fontsize=12)
plt.ylabel('Maximum Delay Times (ns)', fontsize=12)
plt.title('Maximum Non-Linear Delay Times', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# Add custom legend
blue_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Cedars')
red_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='BOIC')
plt.legend(handles=[blue_patch, red_patch], title="Sample Sites")

plt.tight_layout()
plt.savefig('MaxNonLinear.png')
plt.show()

```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Data
samples = ['5A', '5B', '2A', '2B', '1A', '1B', '6B', '6A', '4A']
percentages = [0.24, 0.35, 0.43, 0.55, 0.86, 0.90, 1.80, 1.83, 3.18]
terms_pB = [17.80890483, 10.64218707, 6.592535362, 2.160063122,
            5.721029415, 2.035363251, 1.253916495, 20.01817464, 9.437480219]
terms_sB = [19.67702419, 14.27, 7.551611187, 2.544087847,
            7.839297272, 1.972924522, 2.975833198, 16.92934187, 4.61000944]

# X positions for samples
x_positions = range(len(samples))

# Samples to be colored blue and red
blue_samples = ['5A', '5B', '6B', '6A', '4A']
red_samples = ['2A', '2B', '1A', '1B']

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot for P-wave values
for i, sample in enumerate(samples):
    color = 'red' if sample in red_samples else 'blue'
    plt.scatter(i, terms_pB[i], color=color, marker='x', s=100, edgecolor='black')

# Scatter plot for S-wave values
for i, sample in enumerate(samples):
    color = 'red' if sample in red_samples else 'blue'
    plt.scatter(i, terms_sB[i], color=color, marker='s', s=100, edgecolor='black')

# Add grid lines
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Add labels and title
plt.xlabel('Samples (Ordered from Lowest to Highest Carbonation)')
plt.ylabel('Anisotropy Values (%)')
plt.title('P- and S-Wave Anisotropies')

# Custom x-axis tick labels with sample + percentage
xtick_labels = [f'{sample}\n({pct:.2f}%)' for sample, pct in zip(samples, percentages)]
plt.xticks(x_positions, xtick_labels, fontsize=10)

# Legend handles
handles = [
    Line2D([0], [0], marker='x', color='w', markerfacecolor='blue', markeredgecolor='black',
           markersize=10, label='Cedars P-wave'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='blue', markeredgecolor='black',
           markersize=10, label='Cedars S-wave'),
    Line2D([0], [0], marker='x', color='w', markerfacecolor='red', markeredgecolor='black',
           markersize=10, label='BOIC P-wave'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='red', markeredgecolor='black',
           markersize=10, label='BOIC S-wave')
]

# Add legend
plt.legend(handles=handles)

plt.tight_layout()
plt.savefig('aniRIGHT.png')
plt.show()
```
